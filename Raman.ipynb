```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
df = pd.read_csv(r"8c.txt",sep="\t",header=None)

# Asumo que las columnas son las dos únicas
x = df.iloc[:, 0]
y = df.iloc[:, 1]

plt.figure(figsize=(8,6))
plt.plot(x, y)
plt.xlabel("Longitud de onda (cm**-1)")
plt.ylabel("Intensidad (u.a.)")
plt.title("Dispersión de datos")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ruta al archivo
archivo = r"8c.txt"

# Cargar datos
df = pd.read_csv(archivo, sep="\t", header=None)

# Asumo que tienes dos columnas: x e y
x = df.iloc[:, 0]
y = df.iloc[:, 1]

# Nombre base del archivo (sin extensión)
nombre_base = os.path.splitext(os.path.basename(archivo))[0]

# Crear figura explícita
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y)

# Líneas verticales
ax.axvline(x=1580, color="red", linestyle="--", label="G")
ax.axvline(x=2670, color="blue", linestyle="--", label="G'")

# Etiquetas
ax.set_xlabel("Longitud de onda (cm$^{-1})$",fontsize=12)
ax.set_ylabel("Intensidad (u.a.)",fontsize=12)
ax.set_title("Espectro Raman")
ax.legend()
ax.grid(True)

# Guardar con el mismo nombre que el archivo
salida = os.path.join(os.path.dirname(archivo), f"{nombre_base}.png")
fig.savefig(salida, dpi=300, bbox_inches="tight")
plt.show()

plt.close(fig)  # ahora sí cerramos la figura
print(f"Gráfica guardada en: {salida}")
```
