```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ruta al archivo
archivo = r"C:\Users\theye\Downloads\PABLO U_532_1200_100x_0_75 NIR_100_300_100%.txt"

# Cargar datos
df = pd.read_csv(archivo, sep="\t", header=None)

# Asumo que tienes dos columnas: x e y
x = df.iloc[:, 0]
y = df.iloc[:, 1]

# Nombre base del archivo (sin extensin)
nombre_base = os.path.splitext(os.path.basename(archivo))[0]

# Crear figura explcita
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y)

# Lneas verticales
ax.axvline(x=1580, color="red", linestyle="--", label="G")
ax.axvline(x=2720, color="blue", linestyle="--", label="G'")

# Etiquetas
ax.set_xlabel("Longitud de onda (cm$^{-1})$")
ax.set_ylabel("Intensidad (u.a.)")
ax.set_title("Espectro Raman")
ax.legend()
ax.grid(True)

# Guardar con el mismo nombre que el archivo
salida = os.path.join(os.path.dirname(archivo), f"{nombre_base}.png")
fig.savefig(salida, dpi=300, bbox_inches="tight")
plt.show()

plt.close(fig)  # ahora s cerramos la figura
print(f"Grfica guardada en: {salida}")
```
